```python
import django 
import os
import requests

os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'blog.settings')
os.environ.setdefault('DJANGO_ALLOW_ASYNC_UNSAFE', 'true')
django.setup()
```

```python
from article.models import Post
Post.objects.all()
```

```python
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)
data = response.json()
```

```python
from django.contrib.auth import get_user_model
User = get_user_model()
user = User.objects.get(id=1)
```

```python
data[0]
```

```python
Post.objects.all().delete()
```

```python
post_cid_id_dict = {}
for post in data:
   p=Post.objects.create(title=post['title'], body=post['body'],slug=post['id'],author=user)
   post_cid_id_dict[post['id']] = p.id
    
```

```python
url = "https://jsonplaceholder.typicode.com/comments"
response = requests.get(url)
data_comment = response.json()
```

```python
from article.models import Comment, Like
Comment.objects.all()
```

```python
data_comment[0]
```

```python
Comment.objects.all().delete()
```

```python
for comment in data_comment:
    Comment.objects.create(author_comment=user, body=comment['body'], 
                           post_id=post_cid_id_dict[comment['postId']])
    
```
